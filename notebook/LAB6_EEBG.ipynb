Now, let's use what we learned from the last lab to improve these three graphs.
To start, rerun the following commmands we used last lab to modify the 'titanic' dataset.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

lab3_data = sns.load_dataset('titanic')
lab3_data.loc[(lab3_data['who'] == 'child') & (lab3_data['sex'] == 'male'), 'who'] = 'boy'
lab3_data.loc[(lab3_data['who'] == 'child') & (lab3_data['sex'] == 'female'), 'who'] = 'girl'
lab3_data.loc[lab3_data['survived'] == 0, 'survived_str'] = 'no'
lab3_data.loc[lab3_data['survived'] == 1,  'survived_str'] = 'yes'
```

Next, we will set the style of the plots to '*white*' and the context to '*paper*' using the following:

```python
sns.set_style("white")
sns.set_context("paper")
```

Then, to maintain a consistant color-scheme, we will save the following Python dictionary under a desired name:

```python
veryFunColors = {'man':'blue','woman':'tab:orange','boy':'cornflowerblue','girl':'sandybrown'}
```

###**Plot #1**

Now we will remake the plot with the color palette we have set before (in my case, veryFunColors).

```python
sns.barplot(data=lab3_data, x='class', y='survived', hue='who', capsize=0.2, ci=90, palette=veryFunColors)
```

Next, to better contrast between the richest and poorest individuals on the Titanic, let's remove the 'Second' class category by using a custom x-axis order.

```python
sns.barplot(data=lab3_data, x='class', y='survived', hue='who', capsize=0.2, ci=90, palette=veryFunColors, order=['First', 'Third'])
```

Let's now set the plot's xlabel to 'Boarding Class' and ylabel to 'Mean survival 'rate'.

```python
plot_1=sns.barplot(data=lab3_data, x='class', y='survived', hue='who', capsize=0.2, ci=90, palette=veryFunColors, order=['First', 'Third'])
plot_1.set_xlabel("Boarding Class")
plot_1.set_ylabel ("Mean survival rate")
```

Finally, we will despine this plot with the command below:

```python
plot_1=sns.barplot(data=lab3_data, x='class', y='survived', hue='who', capsize=0.2, ci=90, palette=veryFunColors, order=['First', 'Third'])
plot_1.set_xlabel("Boarding Class")
plot_1.set_ylabel ("Mean survival rate")
sns.despine()
```

###**Plot #2**

For this one, we will be combining the Box Plot and Strip Plot that were produced in the previous lab.

To start, let us produce a colorless version of the previous boxplot as follows:

```python
sns.boxplot(data=lab3_data, x='survived_str', y='age', hue='who', boxprops={'edgecolor':'black','facecolor':(1,1,1,0)})
```

Finally, we will finish the graph by performing the following:
*   Pass the *zorder=0* argument to *sns.stripplot*.
*   Pass the *zorder=5* argument to *sns.boxplot*.
*   Set the *xlabel* to 'Survived' and *ylabel* to 'Age'.
*   Add the command *sns.despine()*.


```python
plot_2=sns.boxplot(data=lab3_data, x='survived_str', y='age', hue='who', boxprops={'edgecolor':'black','facecolor':(1,1,1,0)}, zorder=5)
plot_2=sns.stripplot(data=lab3_data, x='survived_str', y='age', hue='who', jitter=0.3, alpha=0.5, dodge='true', palette=veryFunColors, zorder=0)
plot_2.set_xlabel("Survived")
plot_2.set_ylabel ("Age")
handles, labels = plot_2.get_legend_handles_labels()
plt.legend(handles[4:8], labels[4:8], title='who')
sns.despine()
plt.show()
```

###**Plot #4**

So far the plots I've made are improvements of those from Lab 5. For the remaining two graphs, however, we will once again investigate the passengers of the Titanic, this time with more visualization methods in our toolkit.

To start, let's make a scatter plot displaying 'age' vs. 'fare'. This will provide some insight into how much different groups of passengers paid in 1912 to board what would become a tragic cruise.

The specifications for this graph are as follows:
*  Color datapoints by the 'class' of the individual.
*  Set the *xlabel* to 'Age' and *ylabel* to 'Fare'.
*  Despine the plot.

```python
plot_4 = sns.scatterplot(data=lab3_data, x='age', y='fare', hue='class')
plot_4.set_xlabel("Age")
plot_4.set_ylabel ("Fare")
handles, labels = plot_4.get_legend_handles_labels()
sns.despine()
plt.show()
```

From this, we can see that the variation in what First Class passengeres paid appears much greater than variation within the Second Class and Third Class groups.

Additionally, we can see two outlier individuals in their thirties who paid around 500 for their fare. According to [this website](https://www.in2013dollars.com/uk/inflation/1912?amount=500) plus Google's currency converter, this would be around $80,000 today.

###**Plot #5**

While the scatter plot gave us a general idea what passengers of different ages and boarding classes paid, it was unable to provide us much more demographic-specific information beyond that.

While we could make a box+strip plot or violin plot for each combination of boy/girl/man/woman and first/second/third classes, the resulting graph would probably look very busy.

So instead, let's make use a heatplot.

```python
lab3_data_subset=lab3_data[['class','who','fare']]
lab3_data_subset=lab3_data_subset.dropna(how='any')
lab3_data_subsetAverage=lab3_data_subset.groupby(['class','who']).mean()
lab3_data_subsetAverage=lab3_data_subsetAverage.reset_index()
lab3_data_subsetAverage.head()
```

Next, convert this categorical data into matrix data with the index set as class, columns as who, and values as fare.

```python
matrix_data = lab3_data_subsetAverage.pivot(index='class', columns='who', values='fare')
matrix_data.head()
```

Now we can visualize the heatmap plot.

```python
axes5 = sns.heatmap(data=matrix_data, annot=True,fmt='.2f',cmap='coolwarm')
for text in axes5.texts: text.set_text('' + text.get_text())
axes5.set(xlabel='Demographic', ylabel='Boarding Class', xticklabels=['boy','girl','man','woman'])
```

As we did with the scatter plot, we can see that Second Class and Third Class passengers generally paid around the same amount while individuals in First Class had relatively higher variation in their fares. However, with the heatmap we can now clearly see that within the First Class, girls paid the highest average fare while First Class men paid the least.

Furthermore, we can obtain numerical values for each category directly from the plot itself. This lends itself well toward converting fares from pound sterling in 1912 to their modern-day value.
